### `Substituição do Linkert`

```python
def ordinal_encoder(termo):
    if termo == 'Molto basso' or termo == 'Very low':
        termo = 1
    if termo == 'Basso' or termo == 'Low':
        termo = 2
    if termo == 'Neutro' or termo == 'Neutral':
        termo = 3
    if termo == 'Alto' or termo == 'High':
        termo = 4
    if termo == 'Molto alto' or termo == 'Very High' or termo == 'Very high':
        termo = 5
    if termo == 'Sim' or termo == 'Yes':
        termo = 1
    if termo == 'Não' or termo == 'No':
        termo = 1
    return(termo)
```

```python
cont3=0
for coluna in it:
    it.iloc[:,cont3] = it.iloc[:,cont3].apply(ordinal_encoder)
    cont3+=1
```

```python
def encode(termo):
   termo = str(termo)
   if termo == "Settore secondario (comporta la trasformazione di materie prime in beni)":
      termo = 'secondario'
   if termo == "Settore terziario (coinvolge il commercio e la fornitura di servizi ai consumatori e / o alle imprese)" or termo == 'Tertiary sector (involves trade and the provision of services to consumers and /or companies), Quaternary Sector (involves intellectual activities, such as information generation and sharing)' or termo == 'Settore terziario (coinvolge il commercio e la fornitura di servizi ai consumatori e / o alle imprese), Commercializzazione impianti di riciclo':
      termo = 'terziario'
   if termo == "Settore quaternario (coinvolge attività intellettuali, come la generazione e la condivisione di informazioni)":
      termo = 'quaternario'
   if termo == "Settore quinario (comprende servizi senza scopo di lucro quali sanità, istruzione, cultura, ricerca (non retribuita), polizia, vigili del fuoco, guardia civile e organizzazioni non governative)." or termo == 'Quinary Sector (involves non-profit services such as health, education, culture, research (unpaid), police, firefighters, civil guard and non-governmental organizations).':
      termo = 'quinario'
   #COMBINADOS
   if termo == 'Settore secondario (comporta la trasformazione di materie prime in beni), Settore terziario (coinvolge il commercio e la fornitura di servizi ai consumatori e / o alle imprese), Settore quaternario (coinvolge attività intellettuali, come la generazione e la condivisione di informazioni)' or termo == 'Settore secondario (comporta la trasformazione di materie prime in beni), Settore terziario (coinvolge il commercio e la fornitura di servizi ai consumatori e / o alle imprese)':
      termo = 'secondario e terziario'
   if termo == 'Settore terziario (coinvolge il commercio e la fornitura di servizi ai consumatori e / o alle imprese), Settore quaternario (coinvolge attività intellettuali, come la generazione e la condivisione di informazioni)':
      termo = 'terziario e quaternario'
   if termo == 'Settore quaternario (coinvolge attività intellettuali, come la generazione e la condivisione di informazioni), Settore quinario (comprende servizi senza scopo di lucro quali sanità, istruzione, cultura, ricerca (non retribuita), polizia, vigili del fuoco, guardia civile e organizzazioni non governative).':
      termo = 'quaternario e quinario'
   return (termo)
```
